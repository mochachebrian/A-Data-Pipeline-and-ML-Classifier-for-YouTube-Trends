## Find average views per category

```python
category_views = df.groupby('category_id')['views'].mean()
print(category_views)
```

## Find total likes by category

```python
likes_by_cat = df.groupby('category_id')['likes'].sum()
print(likes_by_cat)
```

## Sorting: Top 10 most viewed videos

```python
top_videos = df.sort_values(by='views', ascending=False).head(10)
print(top_videos[['title', 'views']])
```
